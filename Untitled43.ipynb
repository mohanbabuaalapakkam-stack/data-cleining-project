# **Reading Data Set**

# **Filling Nan Values**
#1.Filling Unnamed Column Using Ffill()

```python
x=df['Unnamed'].median()
df['Unnamed'].fillna(x,inplace=True)
df
```

```python
x=df['Memory'].median()
df['Memory'].fillna(x,inplace=True)
df
```

```python
x=df['Weight'].mean()
df['Weight'].fillna(x,inplace=True)
df
```

```python
x=df['Price'].mode()
df['Price'].fillna(x,inplace=True)
df
```

# **Removing White Space**

```python
df.columns=df.columns.str.strip()
```

# **Filling Nan Values**
#1.Filling Company Column Using Ffill()

```python
df['Company'].fillna(method='ffill',inplace=True)
df
```

#2.Filling Cpu Column Using bfill()

```python
df['Cpu'].fillna(method='bfill',inplace=True)
df
```

#3.Filling Gpu Column Using ffill()

```python
df['Gpu'].fillna(method='ffill',inplace=True)
df
```

```python
df['Price'].fillna(method='ffill',inplace=True)
df
```

#4.Fillna Inches Column Using ffill()

```python
df['Inches'].fillna(method='ffill',inplace=True)
df
```
